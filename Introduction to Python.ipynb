# Introduction to Python

## What is Python?

Python is an interpreted, high-level, general-purpose programming language. Its design philosophy emphasizes code readability with its use of significant indentation. Its language constructs as well as its object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.

## History of Python:

- Laid foundation in 1980s
- Implementation started in Dec, 1989 by Guido Van Rossum
- In 1991, published code labeled version 0.9.0
- Python 1.0 was released in 1994, with features like Lambda and Map
- Python 2.0 was released in 2008, designed to rectify the fundamentals flaw of the language

## Comments in Python

Comments in Python can be written in two ways:
- Single Line Comments
- Multiline Comments

### Single Line Comments

Single line comments are started with an Hash tag.

e.g.

```python
# This is a single line comment
```

### Multiline Comment

Multiline comments are enclosed in a delimiter (""") at eeach end.

e.g.

```python
"""This is a multiline comment"""
```
