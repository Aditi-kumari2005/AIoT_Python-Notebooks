# Python Variables

## Variable Basics

Variables are:
- Containers 
- Stores Data Values

There is no command for declaring a variable. A variable is created the moment you first assign a value to it.

```python
# Declaring a variable and Assiging value "3" to it.

var_1 = 3
```

print() : We can print the variable's value using "print" function

```python
# Printing value of the variable.

print(var_1)
```

type() : We can also check the variable's type using "type" function

```python
type(var_1)
```

## Declaring Variable with Types 

### Variable Types

We can also state the type of variable we are declaring in python just by that variable name followed by its type and in parenthesis, we give its value

```python
# Example 1 : Declaring a variable of type integer
height = int(10.0)
type(height)
```

```python
# Example 2 : Declaring a variable of type float
var_2 = float(10)
print(var_2)
```

### Variable Naming

Variables can have a short name (a,b,c) or they can also have more explanatory name (age, name, total_volume).

There are certain things that we keep in mind before declaring a variable.

- It should start with a letter or underscore
- It can't start with a number
- It can only contain alpha-numeric characters and underscore in it. ( A-z, 0-9, _ )

## Ways to Define Variables

- Assigning multiple values to multiple variables in one line

```python
var_1, var_2, var_3 = 12,23,34
```

```python
# Printing the variable 
print(var_1)
```

```python
# Printing multiple values in single printing statement
print(var_1, var_2, var_3)
```

We can also assign single values to multiple variables

```python
# Here all the variables have same value 23
fruit_1 = fruit_2 = fruit_3 = 23
```

```python
# Checking type of multiple variable in single print statement
print(type(fruit_1), type(fruit_2), type(fruit_3))
```
